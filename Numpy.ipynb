# Aritmética com Array

```python
dados4 = np.array([[1,2,3], [4,5,6]])
dados4
```

```python
# MULTIPLICAÇÃO - multiplica valor por valor  da mesma posição entre os arrays
dados4 * dados4
```

```python
# SUBTRAÇÃO - subtrai valor por valor da mesma posição entre os arrays
dados4-dados4
```

```python
#  ADIÇÃO - somo cada valor das posições com ele mesmo
dados4+dados4
```

```python
# DIVISÃO - divide cada valor das posições por ele mesmo
dados4/dados4
```

```python
# 1 dividido pela valor de cada uma das posições
1/dados4
```

```python
# Somente realizando operações no meu array, eu não crio um novo array com novos valores
# se eu quiser fazer isso, eu preciso criar um novo array que armazene o valor da operação:

dados5 = dados4 + dados4
dados5
```

```python
# COMPARAÇÃO - compara valor por valor
dados5>dados4
```

# Indexação

```python
dadosseq
```

```python
dadosseq[5]
```

Diferente de uma lista python, um recorte (fatia) de um ndarray sofreará broadcast - é como se fosse uma view de banco de dados.
Com conjuntos imensos de dados isso é uma grande ventagem em relação à perfomance.

```python
# do 3 até uma posição antes da 6
dadosseq[3:6]
```

```python
dadosseq[3:6] = 20
dadosseq
```

```python
fatia_dados = dadosseq[2:4]
fatia_dados
```

```python
fatia_dados[1] = 2000
fatia_dados
```

```python
# Quando eu faço uma fatia do meu array, as alterações feitas na fatia também
# são feitas no array original
dadosseq
```

```python
# : significa todos os valores
fatia_dados[:] = 7
dadosseq
```

# Copiar dados
(para não sofrer broadcast)

```python
fatia2 = dadosseq[2:4].copy()
fatia2
```

```python
fatia2[:] = 8
fatia2
```

# Transposição

Para encontrar a matriz transposta, basta trocar a posição das linhas e colunas da matriz.
A função T do numpy faz isso automaticamente.

```python
dados7 = np.arange(15).reshape((3,5))
dados7
```

```python
dados7.T
```

# Multiplicação de Matrizes com Dots

```python
# MULTIPLICAÇÃO DE MATRIZES
np.dot(dados7.T,dados7)
```
